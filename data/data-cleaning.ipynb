```python
# import libraries
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv("idp_pop_wide.csv")
```

```python
print(df1)
```

```python
df2=pd.read_csv("gps.csv")
```

```python
df2=df2.drop(columns=["ref"])
```

```python
df2=df2.rename(columns={"iso3":"iso"})
```

```python
print(df2)
```

```python
filtered_df2 = df2[df2['iso'].isin(df1['iso'])]
```

```python
print(filtered_df2)
```

```python
# Perform the right join
merged_df = pd.merge(df1, filtered_df2, on='iso', how='left')

print(merged_df)
```

```python
merged_df=pd.merge(df1, df2, on="iso", how="right")
```
